```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#데이터 생성
df = pd.DataFrame(columns=['height', 'weight'])
df.loc[0] = [185,60]
df.loc[1] = [180,60]
df.loc[2] = [185,70]
df.loc[3] = [165,63]
df.loc[4] = [155,68]
df.loc[5] = [170,75]
df.loc[6] = [175,80]
```

```python
#데이터 시각화
sns.lmplot(x='height', y='weight',
           data = df, fit_reg=False,
           scatter_kws={"s":200})
```

**K 평균 군집**

```python
# 데이터 프레잉을 Numpy배열로 변환
data_points = df.values
kmeans = KMeans(n_clusters=3).fit(data_points)

kmeans.cluster_centers_
df['cluster_id'] = kmeans.labels_
```

```python
#k 평균 군집 시각화
sns.lmplot(x='height', y='weight', data=df, fit_reg=False,
           scatter_kws = {'s': 150},
           hue='cluster_id')
```
