```python
from transformers import BertTokenizer
tokenizer = \
  BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
text = "Using Transformers is easy!"
tokenizer(text)
```

**파이토치 기반 Transformer 모델 사용** pt

```python
encoded_input = tokenizer(text, return_tensors = "pt")
```

```python
# BERT 기본 모델 다운로드
from transformers import BertModel
model = BertModel.from_pretrained("BERT-base-uncased")
```

```python
# 다운받은 모델에 토크나이저 출력 전달
output = model(**encoded_input)
```

**BERT 모델의 텐서플로 버젼 호출**

```python
# 모델의 버젼을 구체적으로 지정한다.
# BERT 기본 모델 다운에서, 앞에 TF만 붙이면 됨
from transformers import BertTokenizer, TFBertModel
tokenizer = \
  BertTokenizer.from_pretrained('BERT-base-uncased') #토크나이저 호출
model = TFBertModel.from_pretrained("BERT-base-uncased") #모델 호출
text = "  Using Transformer is easy!"

# 텐서플로 기반 Transformer 모델 사용 'tf'
encoded_input = tokenizer(text, return_tensors='tf')
output = model(**encoded_input)
```

**허깅페이스 제공**

미리 제공되는 파이프라인 사용, 마스크 채우기 작업을 수행한다.

```python
from transformers import pipeline
unmasker = \
  pipeline('fill-mask', model='BERT-base-uncased')
unmasker("The man worked as a [MASK].")
```
